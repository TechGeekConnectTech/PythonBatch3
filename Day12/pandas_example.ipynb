```python
mall_data=pd.read_csv("mall_customers.csv")
mall_data.describe()
```

```python
print(mall_data["Annual Income (k$)"].mean())
print(mall_data["Annual Income (k$)"].max())
print(mall_data["Annual Income (k$)"].min())
print(mall_data["Annual Income (k$)"].std())
print(mall_data["Annual Income (k$)"].var())
print(mall_data["Annual Income (k$)"].mode())
print(mall_data["Annual Income (k$)"].median())
```

```python
#print all customers data where age is greater than 40
mall_data[(mall_data["Annual Income (k$)"]>40) & (mall_data["Age"]>40)]
```

```python
mall_data.sort_values(by=["Annual Income (k$)"] , ascending=True)
```

```python
mall_data.isnull().sum()
```

```python
#fill null value with average value
mall_data[mall_data["Age"] == 0] = mall_data["Age"].mean()
mall_data[mall_data["Annual Income (k$)"] == 0] = mall_data["Annual Income (k$)"].mean()
mall_data[mall_data["Spending Score (1-100)"] == 0] = mall_data["Spending Score (1-100)"].mean()
mall_data.tail(5)
```

```python
mall_data.to_csv("mall_customers_filled.csv", index=False)
```

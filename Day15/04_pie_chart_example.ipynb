```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


student_df= pd.read_csv('D:/PythonBatch3/PythonBatch3/Day15/student_grades.csv')
grade_counts=student_df['Grade'].value_counts().sort_index()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Custom colors for each grade category
plt.pie(grade_counts.values,labels=grade_counts.index,autopct='%1.1f%%',startangle=140,colors=colors)
plt.title('Distribution of Student Grades')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

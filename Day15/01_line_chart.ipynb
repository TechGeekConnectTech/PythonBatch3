```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv('D:/PythonBatch3/PythonBatch3/Day15/stock_market_data.csv')
df['Date']= pd.to_datetime(df['Date'])

tcs_data = df[df['Company'] == 'ICICI'].copy()
tcs_data = tcs_data.sort_values('Date')

plt.figure(figsize=(10, 6))
plt.plot(tcs_data['Date'], tcs_data['Close'], marker='o', linestyle='-', color='black')
plt.title('ICICI Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('icici_closing_prices.png')
plt.show()
print("Line plot saved as 'icici_closing_prices.png'")
```

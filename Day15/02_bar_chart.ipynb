```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('D:/PythonBatch3/PythonBatch3/Day15/sales_data.csv')
plt.figure(figsize=(15, 10))
product_sales=df.groupby('Product')['Profit_Margin'].sum().sort_values(ascending=False)
bars=plt.bar(product_sales.index, product_sales.values, color=["#ba2727",'#66b3ff',"#6fc36f",'#ffcc99','#c2c2f0'])
plt.title('Total Profit Margin by Product', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Total Profit Margin', fontsize=14)
plt.xticks(rotation=45)
```

```python
import matplotlib.pyplot as plt
import pandas as pd


df=pd.read_csv('D:/PythonBatch3/PythonBatch3/Day15/student_grades.csv')
subjects = ['Math', 'Physics', 'English', 'Chemistry', 'Biology']
colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#33FFF5']

for subject,color in zip(subjects,colors):
    subjects_marks=df[df['Subject']==subject]['Marks']
    plt.hist(subjects_marks, bins=10, alpha=0.7, label=subject, color=color)

plt.title('Distribution of Student Marks by Subject')
plt.xlabel('Marks')
plt.ylabel('Number of Students')

    
```

```python
import numpy as np

print("Numpy version:", np.__version__)

arr1=np.array([1,2,3,4,5])
arr2=np.array([[1,2,3],[4,5,6]])
print("1D Array:", arr1)
print("2D Array:\n", arr2)
print("Array Shape:", arr2.shape)
print("Array Data Type:", arr1.dtype)
print("Array Sum:", np.sum(arr1))
print("Array Mean:", np.mean(arr1))
print("Array Standard Deviation:", np.std(arr1))
print("Array Max:", np.max(arr1))
print("Array Min:", np.min(arr1))

zero_array=np.zeros((3,4))
one_array=np.ones((2,5))
print("Zero Array:\n", zero_array)
print("One Array:\n", one_array)

range_array=np.arange(0,20,3)
print("Range Array:", range_array)
```

```python
a=np.array([1,2])
b=np.array([[5,6],[7,8]])

print("Matrix Addition:\n", a+b)
print("Matrix Multiplication:\n", a*b)
print("Matrix Subtraction:\n", a-b)

print("power:\n", np.power(a,2))
```
